```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'[A-Za-z]+')  #do not take numeric value
```

```python
df.URL[0]
```

```python
tokenizer.tokenize(df.URL[0])  #braek the data set into tokenizer 
```

```python
df['text_tokenized']= df.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
df.head()
```

```python
#limitization or steaming operation do on it ,we use steaming 
```

```python
 pip install SnowballStemmer
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
Stemmer = SnowballStemmer('english')
```

```python
df['text_stemmed'] = df['text_tokenized'].map(lambda l: [Stemmer.stem(word) for word in l])
```

```python
df.head() 
```

```python
df['text'] = df['text_stemmed'].map(lambda l: ' '.join(l))
```

```python
good_sites = df[df.Label == 'good']
bad_sites = df[df.Label == 'bad']
```

```python
#now convert this text into vectors
#(using count vectorizer otherwise we go with deeplearning so we use worldtool,comodule)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
features  = cv.fit_transform(df.text)
```

```python
features[:5].toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, df.Label, test_size=0.2)
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
l_model = LogisticRegression()
```

```python
l_model.fit(x_train,y_train)
```

```python
l_model.score(x_test,y_test)
```

```python
l_model.score(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print('\nCLASSIFICATION REPORT\n')
print(classification_report(l_model.predict(x_test),y_test,target_names=['Bad','Good']))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
con_mat =pd.DataFrame(confusion_matrix(l_model.predict(x_test),y_test),
                      columns =['Predicted:Bad','Predicted:Good'],
                      index =['Actual:Bad','Actual:Good'])
```

```python
import seaborn as sns
```

```python
print('\nCONFUSION MATRIX')
plt.figure(figsize=(6,4))
sns.heatmap(con_mat, annot = True,fmt='d', cmap="YlGnBu")
```
